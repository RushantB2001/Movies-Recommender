```python
import ast 
```

```python
def convert (obj):
    if obj is None :
        return []
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres']= movies['genres'].apply(convert)
movies['keywords']= movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
movies 
```

```python
def convert3 (obj):
    if obj is None :
        return []
    L=[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else :
            break
        return L
         
          
```

```python
movies['cast']= movies['cast'].apply(convert3)
```

```python
def fetch_crew (obj):
    if obj is None:
        return []
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
        return L
```

```python
movies['crew']= movies['crew'].apply(fetch_crew)
```

```python
movies['overview']= movies['overview'].astype(str)
movies['overview']= movies['overview'].apply(lambda x: x.split())
```

```python
movies['overview']
```

```python
movies['genres']= movies['genres'].apply(lambda x :[i.replace(" ","")for i in x])
movies['cast']= movies['cast'].apply(lambda x :[i.replace(" ","")for i in x])
movies['crew']= movies['crew'].apply(lambda x :[i.replace(" ","")for i in x])
```

```python
movies['tag']= movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
for index, row in movies.iterrows():
    value = row['tag']
    
    # If the value is a string, convert it to lowercase
    if isinstance(value, str):
        movies.at[index, 'tag'] = value.lower()
    
    # If the value is a list, join its elements into a string
    elif isinstance(value, list):
        movies.at[index, 'tag'] = ', '.join(map(str, value))  # Convert all items in the list to strings
    
    # For other types, leave them unchanged or apply necessary transformations
    else:
        pass 
```

```python
import nltk 
from nltk.stem.porter import PorterStemmer 
```

```python
ps= PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y) 
```

```python
new_movies['tag']= new_movies['tag'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
similarity= cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])), reverse = True, key = lambda x: x[1]) [1:6]
```

```python
def recommend(movie):
    movie_index= new_movies[new_movies['title']==movie].index[0]
    distance= similarity[movie_index]
    movies_list= sorted(list(enumerate(distance)), reverse =True, key = lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_movies.iloc[i[0]].title)
```

```python
recommend('Iron Man')
```
